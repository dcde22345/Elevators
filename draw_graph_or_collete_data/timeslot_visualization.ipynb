```python
import matplotlib.pyplot as plt
import pandas as pd

# 
data = {
    "Time": ["First 5 Minutes", "Last 5 Minutes"],
    "Morning_Elevator": [26, 28],
    "Morning_Stairs": [16, 20],
    "Noon_Elevator": [34, 37],
    "Noon_Stairs": [14, 23],
    "Afternoon_Elevator": [42, 37],
    "Afternoon_Stairs": [13, 41],
}

df = pd.DataFrame(data)

# 
plt.figure(figsize=(12, 6))

#  - 
plt.plot(df["Time"], df["Morning_Elevator"], label="Morning Elevator", marker='o', color='blue')
plt.plot(df["Time"], df["Morning_Stairs"], label="Morning Stairs", marker='o', color='blue', linestyle='--')

#  - 
plt.plot(df["Time"], df["Noon_Elevator"], label="Noon Elevator", marker='o', color='green')
plt.plot(df["Time"], df["Noon_Stairs"], label="Noon Stairs", marker='o', color='green', linestyle='--')

#  - 
plt.plot(df["Time"], df["Afternoon_Elevator"], label="Afternoon Elevator", marker='o', color='red')
plt.plot(df["Time"], df["Afternoon_Stairs"], label="Afternoon Stairs", marker='o', color='red', linestyle='--')

plt.title("Elevator and Stairs Usage on 5/15 (First vs Last 5 Minutes)")
plt.xlabel("Time Period")
plt.ylabel("Number of People")
plt.xticks(rotation=0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
